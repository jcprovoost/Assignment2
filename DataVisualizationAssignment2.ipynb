```python
import pandas as pd
import numpy as np

data = pd.read_csv('Data.csv')
```

```python
from bokeh.io import show, output_notebook, push_notebook
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnDataSource, ColorBar, LinearColorMapper, ImageURL, Triangle, Label
from bokeh.layouts import Row, Column
from bokeh.palettes import RdYlGn
from ipywidgets import interact, IntSlider, Dropdown
```

# The Big Mac Labor Gap?
Blablabla

```python
hover = HoverTool(tooltips=[("Country", "@Country_Name"),("Big Mac price", "@bigMacPrice $"),("Average yearly income", "@gniPerCap $"),("Average yearly working time", "@averageWorkingHours hrs."),("Working time to buy one Big Mac", "@hoursPerBigMac hrs.")])
tools = ["box_select", hover, "reset"]

source = ColumnDataSource(data.groupby("Country_Name").mean())
output_notebook()

p1 = figure(title="People get paid more for the same amount of work",x_axis_label='Mean yearly income per capita ($)', y_axis_label='Hours of work needed for one Big Mac', width=480, height=450, x_range=(-2000,120000), y_range=(0,3), tools=tools)
s1 = p1.scatter("gniPerCap", "hoursPerBigMac", source=source,radius="radius",hover_color="red",fill_alpha=0.2,color="grey")

p2 = figure(title="People get paid more for the same amount of work",x_axis_label='Yearly working time per capita (hrs.)',y_axis_label='Mean hourly income per capita ($)',x_range=(1300,2500),y_range=(-1,76),width=480, height=450, tools=tools)
s2 = p2.scatter("averageWorkingHours", "wagePerHour", source=source,radius=10,hover_color="green",fill_alpha=0.2,color="green")   

for x in data["Country_Name"].unique():
    p1.line("gniPerCap","hoursPerBigMac",source=ColumnDataSource(data[data["Country_Name"]==x].sort_values("year")),line_alpha=0.15,color="green")

def update1(f):
    s1.data_source.data = ColumnDataSource.from_df(data[data["year"] == f])
    push_notebook()

show(Row(p1,p2), notebook_handle=True)

menu1 = IntSlider(min=2000,max=2016,step=1,value=2000,description="Year:")
interact(update1, f=menu1)
```

```python
hover = HoverTool(tooltips=[("Country", "@Country_Name"),("Population", "@population"),("McDonald's locations", "@mcDonaldsAmount"), ("Average yearly income", "@gniPerCap $")])
tools = ["box_select", hover, "reset"]

mcdonalds_per_country = data[data["mcDonaldsPerPopulation"] > 0]
source = ColumnDataSource(mcdonalds_per_country.groupby("Country_Name").min())

p3 = figure(title="Amount of McDonald's grows exponentially with population",x_axis_label='Total population', y_axis_label="Amount of McDonald's locations", x_axis_type="log", y_axis_type="log",x_range=[5e4, 5e9],y_range=[10e-1, 10e3],width=480, height=450, tools=tools)
s3 = p3.scatter("population", "mcDonaldsAmount", source=source,fill_alpha=0.2,color="grey")

p5 = figure(title="The richer a country, the more McDonald's locations",x_axis_label='Mean yearly income per capita ($)', y_axis_label="Amount of McDonald's locations",x_range=(0,99000),y_range=(-1,50),width=480, height=450, tools=[hover])
s5 = p5.scatter("gniPerCap", "mcDonaldsPerPopulation", source=source,fill_alpha=0.2,color="grey")

show(Row(p3,p5), notebook_handle=True)

def update(f):
    s3.data_source.data = ColumnDataSource.from_df(mcdonalds_per_country[mcdonalds_per_country["year"] == f])
    push_notebook()

menu2 = Dropdown(options=[2010, 2015],value=2010,description='Year:')
interact(update, f=menu2)
```

```python
p4 = figure(title="",x_range=mcdonalds_per_country.sort_values("mcDonaldsPerPopulation", ascending=False)["Country_Name"].unique(), y_range=(0,50), plot_width=2000, plot_height=400, outline_line_color=None,tools=[hover])
s4 = p4.vbar(x="Country_Name", top="mcDonaldsPerPopulation", color="lightcoral", width=0.5, source=source)
p4.xaxis.major_label_orientation = "vertical"

show(p4, notebook_handle=True)

def update(f):
    s4.data_source.data = ColumnDataSource.from_df(mcdonalds_per_country[mcdonalds_per_country["year"] == f])
    push_notebook()

menu2 = Dropdown(options=[2010, 2015],value=2010,description='Year:')
interact(update, f=menu2)
```
